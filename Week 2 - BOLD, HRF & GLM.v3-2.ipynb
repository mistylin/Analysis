## From Neural Impulses To fMRI BOLD response

BOLD signals can tell us something about neural firing, as you have read in the Logothetis 2008 review article. But how do we connect BOLD signals we measure to neural firing? For this, we use the HRF. This is the BOLD response we expect for an impulse (really short burst) of neural firing. There are a couple of descriptions of these, two are defined and plotted below. They differ in whether they have an undershoot after the initial peak, and are called 'gamma' because they relate to the gamma probability distribution from statistics. This function is often used to describe right-skewed shapes.

```python
def double_gamma(x, a1 = 6, a2 = 12, b1 = 0.9, b2 = 0.9, c = 0.35):
	d1 = a1 * b1
	d2 = a2 * b2
	return np.array([(t/(d1))**a1 * np.exp(-(t-d1)/b1) - c*(t/(d2))**a2 * np.exp(-(t-d2)/b2) for t in x])

def single_gamma(x, a = 6, b = 0.9):
	d = a * b
	return (x/d)**a * np.exp(-(x-d)/b)
```

```python
# we evaluate the HRF functions on a thousand timepoints between 0 and 25 s
timepoints_to_evaluate = np.linspace(0,25,1000)

# we plot the two hrfs together, and their difference
f = plt.figure(figsize = (10,4))
s = f.add_subplot(111)
plt.plot(timepoints_to_evaluate, single_gamma(timepoints_to_evaluate, a = 6, b = 0.9),
         'r', lw = 4, alpha = 0.4)
plt.plot(timepoints_to_evaluate, double_gamma(timepoints_to_evaluate, a1 = 6, a2 = 12, b1 = 0.9, b2 = 0.9, c = 0.35),
         'b', lw = 4, alpha = 0.4)
plt.plot(timepoints_to_evaluate, -0.35 * single_gamma(timepoints_to_evaluate, a = 12, b = 0.9),
         'g--', lw = 4, alpha = 0.4)
plt.legend(['single gamma, no undershoot', 'double gamma with undershoot', 'the double-gamma undershoot'])
plt.xlabel('time [s]')
s.set_ylim([-0.4,1.2])
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.ylabel('BOLD response strength')
plt.title('Single-gamma and Double-gamma BOLD response HRFs')
```

From the **single_gamma** and **double_gamma** functions and their arguments it's clear that the double gamma is the sum of two single gamma's, right? ## From Neural Signals in your experiment to BOLD responses

Neural events in your experiment aren't necessarily very short. They can vary in duration and intensity, so how do we relate the BOLD responses we measure to the 'neural activations' we want to say something about? The idea is that the signal we record using fMRI, the BOLD response, is neural activations, 'filtered through" the HRF. The mathematical operation that does the 'filtering' is called **convolution**.

Here, we will illustrate the concept of convolution.

Convolution is a mathematical operation that is intuitively best understood as follows: if you convolve one function with another, you take the value of the first function, and multiply it with every 'time-point' of the second in a 'sliding' fashion. This quantifies the area the two functions have in common at each timepoint of the 'sliding' operation. 

Here's a movie that describes what this is, taken from the wikipedia entry on <a href="https://en.wikipedia.org/wiki/Convolution">convolution</a>.

<img src="https://upload.wikimedia.org/wikipedia/commons/b/b9/Convolution_of_spiky_function_with_box2.gif">

Here, think of the red 'box' as neural activity, and the sloping blue shape as a weird-looking HRF. What is clear is that the two are sliding over one another, producing the convolution of the two. The standard way of denoting convolution is by a *.

For the purposes of this course, best is to take this intuitive understanding from the following graphical description in terms of neural and BOLD responses.

## The haemodynamic transformation is linear 
The reading material for this week (the Boynton 2012 article on blackboard) shows a very important characteristic of the BOLD response, and its relation with neural firing. The transformation between neural and BOLD signals is well-described as a linear system, meaning one can reproduce the BOLD response by taking a neural signal and convolving it with the HRF. 
So, the neural and BOLD responses are related to one another in a linear way. This means that summing already convolved BOLD responses to different neural events is identical to combining the neural events, and then performing a convolution with the HRF. Below, we illustrate these characteristics of the BOLD response.

Read through the code below, and try to understand, from the figures why these properties of the haemodynamic transformation are important!

```python
# we create an input signal and hrf 
# on the same time axis. 
f = plt.figure(figsize = (10,8))

# first, we create a neural signal, which is a step function
input = np.concatenate([np.zeros(200), np.ones(10), np.zeros(790)])

# and, we need an hrf
hrf = double_gamma(timepoints_to_evaluate, a1 = 6, a2 = 12, b1 = 0.9, b2 = 0.9, c = 0.35)
# hrf = hrf/hrf.max()

# the bold response we expect is the convolution of the two.
# because we do 'full' deconvolution, the resulting array becomes longer by the size of the hrf. 
# we don't want that, and thus we take only those convolved timepoints that are inside the scope of 
# the original data's time (see [:timepoints_to_evaluate.shape[0]]).
impulse_response = np.convolve(input, hrf, 'full')[:timepoints_to_evaluate.shape[0]]

s = f.add_subplot(311)
plt.title('Convolution of neural signal with HRF')
plt.plot(timepoints_to_evaluate-5.0, impulse_response,
         'g--', lw = 3, alpha = 0.4)
plt.plot(timepoints_to_evaluate-5.0, input*10,
         'g', lw = 2, alpha = 1)
plt.legend(['BOLD response', 'Neural signal'])
plt.xlabel('time [s]')
s.set_ylim([-40,120])
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.ylabel('BOLD response strength')

s = f.add_subplot(312)
plt.title('Convolution of 10 neural signals with HRF, summed together')
for x in range(1, 11):
    plt.plot(timepoints_to_evaluate-5.0, impulse_response * x,
         'g--', lw = 3, alpha = 1.0 * x/10.0)
    plt.plot(timepoints_to_evaluate-5.0, input*10*x,
         'g', lw = 1, alpha = 1.0 * x/10.0)
plt.legend(['BOLD response', 'Neural signal'])
plt.xlabel('time [s]')
s.set_ylim([-40,120])
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.ylabel('BOLD response strength')

# now, let's see what happens if we put in a stronger neural signal
input = np.concatenate([np.zeros(200), np.ones(10)* 10, np.zeros(790)])
impulse_response = np.convolve(input, hrf, 'full')[:timepoints_to_evaluate.shape[0]]

s = f.add_subplot(313)
plt.title('Convolution of 10x stronger neural signal with HRF')
plt.plot(timepoints_to_evaluate-5.0, impulse_response,
         'g--', lw = 3, alpha = 1)
plt.plot(timepoints_to_evaluate-5.0, input*10,
         'g', lw = 2, alpha = 1)
plt.legend(['BOLD response', 'Neural signal'])
plt.xlabel('time [s]')
s.set_ylim([-40,120])
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.ylabel('BOLD response strength')

plt.tight_layout()
```

Note that the last two signals are the same! 

It doesn't matter whether you sum the neural responses, or the BOLD responses. 

```python
# now, let's see what happens if we put in a longer-lasting neural signal
f = plt.figure(figsize = (10,8))
timepoints_to_evaluate = np.linspace(0,100,1000)
hrf = double_gamma(timepoints_to_evaluate, a1 = 6, a2 = 12, b1 = 0.9, b2 = 0.9, c = 0.35)

input = np.concatenate([np.zeros(50), np.ones(50), np.zeros(900)])
impulse_response = np.convolve(input, hrf, 'full')[:timepoints_to_evaluate.shape[0]]

s = f.add_subplot(131)
plt.title('Convolution of \n brief neural signal \n with HRF')
plt.plot(timepoints_to_evaluate-5.0, impulse_response,
         'b--', lw = 3, alpha = 0.4)
plt.plot(timepoints_to_evaluate-5.0, input*10,
         'b', lw = 2, alpha = 1)
plt.legend(['BOLD response', 'Neural signal'], loc = 1 )
plt.xlabel('time [s]')
s.set_ylim([-20,60])
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.ylabel('BOLD response strength')


s = f.add_subplot(132)
plt.title('Convolution of \n many brief neural signals, \n separated in time \n with the HRF, and their sum')
summed_response = np.zeros_like(impulse_response)
for x in range(0,10):
    summed_response += np.roll(impulse_response, 50 * x)
    plt.plot(timepoints_to_evaluate-5.0, np.roll(impulse_response, 50 * x),
         'b--', lw = 1.5, alpha = 1.0 * x/10.0)
    plt.plot(timepoints_to_evaluate-5.0, np.roll(input*10, 50 * x),
         'b', lw = 2, alpha = 1.0 * x/10.0)
plt.plot(timepoints_to_evaluate-5.0, summed_response,
         'b', lw = 3, alpha = 1)
# plt.legend(['BOLD response', 'Neural signal'], loc = 0 )
plt.xlabel('time [s]')
s.set_ylim([-20,60])
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.ylabel('BOLD response strength')

# shifting the responses and inputs, and then summing, is the same as summing the inputs and then creating the response
# by convolution

input = np.concatenate([np.zeros(50), np.ones(450), np.zeros(500)])
impulse_response = np.convolve(input, hrf, 'full')[:timepoints_to_evaluate.shape[0]]

s = f.add_subplot(133)
plt.title('Convolution of \n longer-lasting neural signal \n with HRF')
plt.plot(timepoints_to_evaluate-5.0, impulse_response,
         'b--', lw = 3, alpha = 0.4)
plt.plot(timepoints_to_evaluate-5.0, input*10,
         'b', lw = 2, alpha = 1)
plt.legend(['BOLD response', 'Neural signal'], loc = 1 )
plt.xlabel('time [s]')
s.set_ylim([-20,60])

plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.ylabel('BOLD response strength')

plt.tight_layout()
```

Again, it doesn't matter if you first add the neural signals together and convolve them (rightmost figure), or if you convolve the neural signals first and then add the convolved signals (middle figure). 

But imagine what would happen if this 'linearity' was not the case. It would mean that it would be extremely hard to find the neural response from the BOLD response, because it might matter when the last stimulus occurred, how strong it was, etc. etc. fMRI analysis would be hell! But because the haemodynamic transformation is a linear one, we can use assumptions in our analyses, opening up the analysis methods that we'll be talking about later in this notebook. 

## Linear Regression - simplest case

We're now going to take a step sideways, and look at the type of analysis is the standard for fMRI analysis. It's based on linear regression. You may have some vague recollection of this statistical method from your earlier statistics course(s).

Linear regression is a type of analysis that allows you to link two variables to one another, by means of a linear relationship. Let's first create some data, to illustrate what type of data is amenable to linear regression analysis. 

```python
nr_of_points = 100
x_points = 4 * np.random.randn(nr_of_points) + 1.25
y_points = -3.245 * x_points + 4.3 + np.random.randn(nr_of_points)

f = plt.figure(figsize = (6,6))
f.add_subplot(111)
plt.title('Scatterplot')

plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

plt.plot(x_points, y_points, 'ro')
```

One way to describe these data, and the one we will use, is by defining a linear relation between them, and finding the parameters that best describe this relation. 
A linear function of one variable is described by the equation

$ y = a \cdot x + b $, 

where $a$ is the slope of the relation between $x$ and $y$, and $b$ is the offset, or intercept. We can use linear regression to find the parameters **$a$** and **$b$** - which in the above example have the value of -3.245 and 4.3, respectively. Can you find those in the code?

Note that we did add noise to the data, in the form of normally distributed random numbers. Let's try that regression. We also say that we fit the parameters. 

```python
# we perform linear regression using least squares fitting
# please, have a look at the help of np.linalg.lstsq help independently.
betas, _sse, _rank, _svs = np.linalg.lstsq(np.array([x_points, np.ones(y_points.shape[0])]).T, y_points )

slope, intercept = betas
```

Please, have a look at the input to the function `lstsq`. What we put in is the data, `y_points`, and a numpy array which we call the 'coefficient matrix' or **design matrix**. This numpy arrays has two columns, one of which is the `x_points` -- this column will allow us to find the slope parameter $a$ of the relation between $x$ and $y$. But notice that we've also added a `np.ones` column. These ones model the intercept $b$ of the fit. 

Notice also that the variable that is used to contain the resulting slope and intercept is called `betas`, for **beta weights**. This is the standard term, and it will recur several times in this course. 

```python
f = plt.figure(figsize = (6,6))
f.add_subplot(111)
plt.title('Scatterplot with fit \n slope $a$ = %2.2f, intercept $b$ = %2.2f'%(betas[0], betas[1]))

plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

fitted_x_points = np.linspace(-10,15,100)
fitted_y_points = slope * fitted_x_points + intercept

plt.plot(x_points, y_points, 'o', alpha = 0.5, mec = 'r', mew = 1.5, color = 'w')
plt.plot(fitted_x_points, fitted_y_points, 'k--', lw = 2.5, alpha = 0.5)
```

Can you see the intercept and slope in the black dashed line in the figure? That is, what do the two mean, in this graph? And, can you already imagine how you would use this type of analysis in fMRI? Just think about the situation in which your experiment manipulates $x$, while you measure $y$. 

## Multiple Linear Regression: General Linear Model
The above example of fitting a linear relation between variables can be extended to multiple variables that try to explain (the variance of) another. The variables that try to explain are (trivially) called 'explanatory variables'.
We fit multiple explanatory variables to the data in what is called a **general linear model** approach.

The same piece of code as we used above can be made to fit a general linear model with more explanatory variables than just one; the function `lstsq` doesn't care how many explanatory variables there are. Let's try that out by adding another explanatory variable, with its own slope parameter. The equation becomes

$ y = a \cdot x1 + b \cdot x2 + c + N(\mu, \sigma) $, 

where $N(\mu, \sigma)$ is normally distributed noise to make things look more 'real'.
What does this mean? Now, each value of y depends not just on a single x, but on two xs. Let's see if we can quantify the different roles of the two xs...

```python
# a linear equation of multiple parameters:
# the parameters are:
a, b, c = 1.5, -2.4, 3.002
# where a and b are slopes, and c is the intercept. 
# the variables with the a and b parameters are x1 and x2

x1, x2 = np.random.randn(100), np.random.randn(100)
x = np.array([x1,x2])

# creating values of y according to the equation above
y = x1 * a + x2 * b + c + np.random.randn(100)

f = plt.figure(figsize = (6,6))
f.add_subplot(111)
plt.title('Scatterplot')

plt.xlabel('X, 1 and 2')
plt.ylabel('Y')
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)
plt.axvline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

plt.plot(x1, y, 'o', alpha = 0.5, mec = 'r', mew = 1.5, color = 'w')
plt.plot(x2, y, 'o', alpha = 0.5, mec = 'b', mew = 1.5, color = 'w')
```

So, plotting them separately, we see that values of `y` depend on both `x1` and `x2`. But, we really need to look at this in 3 dimensions, because the values of `y` depend on both `x1` and `x2` at the same time:

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(y, x1, x2, c='g', marker='o')
plt.xlabel('X1')
plt.ylabel('X2')
for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
```

Please, play around with this plot (if your computer allows it)

You will see that the data are a plane in three dimensions, with some noise added. You also notice that the amount of noise on top of the plane is far less than the noise on top of each of the above scatter plots, separately. 

```python
betas, _sse, _r, _svs = np.linalg.lstsq(np.array([x1, x2, np.ones(y.shape[0])]).T, y )

print('-----------Expected values:-------------')
print(a, b, c)

print('-----------Actual values:-------------')
print(betas)


# meshgrid is a convenience function used to create a nice rectangular set of x and y coordinates.
x1_for_plot, x2_for_plot = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2.5, 2.5, 20))

# We can use the above equation and the best-fitting beta values
# to see what the fitted values for y are.
y_for_plot = x1_for_plot * betas[0] + x2_for_plot * betas[1] + betas[2] 

# now we also want to plot these
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y, c='k', marker='v')

surf = ax.plot_surface(x1_for_plot, x2_for_plot, y_for_plot,
                       linewidth=1, antialiased=True, alpha = 0.3)

plt.xlabel('X1')
plt.ylabel('X2')

```

# General Linear Model in fMRI

### Or, how do we use this stuff in fMRI? 

The funny thing is, if you understand the above manipulation, you've understood the basics of most fMRI analyses that have ever been done. Most fMRI analyses consist of running the analysis you ran above, but then for tens of thousands of voxels at the same time. It's also likely that researchers use more than two variables playing the role of `x1` and `x2` in the above example.

To show you how all of this goes into an fMRI experiment, we'll take the following approach. We'll **create** a set of data for three imaginary, 'model', voxels. One of our voxels responds predominantly to visual information, another predominantly to auditory information, and the last is a 'multimodal' voxel, which responds to both.
First, we create 'neural signals', and then, using the linearity of the haemodynamic transformation, we create according 'BOLD' responses. We also add noise, both to the neural signal and to the BOLD response. 

This type of endeavour, creating the data that you expect to find, is a very helpful method for complicated situations, such as fMRI analysis, because you know what you put in - and that's exactly what you'll have to be able to get out! 

Please have a look at the following code, where we model two types of events (auditory and visual) that may occur during our imaginary experiment, and three possible responses there could be to these events.

```python
neural_signal_1 = np.concatenate([np.roll(np.r_[np.ones(100),np.zeros(500)], np.random.randint(-200,200, 1)) 
                                  for x in range(10)])
neural_signal_2 = np.concatenate([np.roll(np.r_[np.zeros(300),np.ones(100),np.zeros(200)], np.random.randint(-200,200, 1)) 
                                  for x in range(10)])

f = plt.figure(figsize = (10,8))
s = f.add_subplot(411)
plt.title('Neural activity / Stimulus Presentation')

plt.plot(np.arange(neural_signal_1.shape[0])/20.0, neural_signal_1, lw = 2.5)
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, neural_signal_2, lw = 2.5)

plt.xlabel('time [s]')
s.set_ylim([-0.5,1.35])
s.set_xlim([-5,305])

plt.legend(['Visual', 'Auditory'])

# and, we need an hrf
timepoints_to_evaluate = np.linspace(0,25,500) # ~20 Hz sampling in this simulation
hrf = double_gamma(timepoints_to_evaluate, a1 = 6, a2 = 12, b1 = 0.9, b2 = 0.9, c = 0.35)

BOLD_timecourse_1 = np.convolve(neural_signal_1, hrf, 'full')[:neural_signal_1.shape[0]]
BOLD_timecourse_2 = np.convolve(neural_signal_2, hrf, 'full')[:neural_signal_2.shape[0]]

s = f.add_subplot(412)
plt.title('Model BOLD Signals Without Noise')

plt.plot(np.arange(neural_signal_1.shape[0])/20.0, 
         BOLD_timecourse_1, lw = 2.5)
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         BOLD_timecourse_2, lw = 2.5)
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

plt.xlabel('time [s]')
plt.legend(['Visual', 'Auditory'])

#
# now we know what visual and auditory signals should look like in our simulated experiment,
# we can create our three voxels. 
# They will have Audio-visual weights of: [0.0, 3], [3.75, 0.0], [4.0, 4.0]
# And, as in real life, this simulation of the neural_signal will be noisy.
#

noisy_neural_signal_1 = neural_signal_1 + np.random.randn(neural_signal_1.shape[0])
noisy_neural_signal_2 = neural_signal_2 + np.random.randn(neural_signal_2.shape[0])

# Question:
# does it matter whether we convolve before or after the multiplications?

visual_voxel_timecourse = np.convolve(3.0 * noisy_neural_signal_1, hrf, 'full')[:noisy_neural_signal_1.shape[0]]
auditory_voxel_timecourse = np.convolve(3.75 * noisy_neural_signal_2, hrf, 'full')[:noisy_neural_signal_1.shape[0]]
multimodal_voxel_timecourse = np.convolve(4.0 * noisy_neural_signal_2 + 4.0 * noisy_neural_signal_1, hrf, 'full')[:noisy_neural_signal_1.shape[0]]

s = f.add_subplot(413)
plt.title('3 Voxel BOLD Signals With Neural Noise')

plt.plot(np.arange(neural_signal_1.shape[0])/20.0, 
         visual_voxel_timecourse/visual_voxel_timecourse.max(), lw = 2.5, color = 'm')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         auditory_voxel_timecourse/auditory_voxel_timecourse.max(), lw = 2.5, color = 'c')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         multimodal_voxel_timecourse/multimodal_voxel_timecourse.max(), 'k--', lw = 2.5, alpha = 0.7)


plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

plt.xlabel('time [s]')
plt.legend(['Visual Voxel', 'Auditory Voxel', 'Multimodal Voxel'])

#
# The three voxels have different levels of measurement noise
# Question: what are the differences between the measurement noise and the neural noise?
#
random_intercepts = 1000 * np.random.randn(3)

visual_voxel_timecourse_N = random_intercepts[0] + visual_voxel_timecourse + visual_voxel_timecourse.max() * 0.08 * np.random.randn(neural_signal_1.shape[0])
auditory_voxel_timecourse_N = random_intercepts[1] + auditory_voxel_timecourse + auditory_voxel_timecourse.max() * 0.05 * np.random.randn(neural_signal_1.shape[0])
multimodal_voxel_timecourse_N = random_intercepts[2] + multimodal_voxel_timecourse + multimodal_voxel_timecourse.max() * 0.15 * np.random.randn(neural_signal_1.shape[0])

s = f.add_subplot(414)
plt.title('3 Voxel BOLD Signals With Neural and Scanner Noise')

plt.plot(np.arange(neural_signal_1.shape[0])/20.0, 
         visual_voxel_timecourse_N, lw = 0.5, color = 'm')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         auditory_voxel_timecourse_N, lw = 0.5, color = 'c')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         multimodal_voxel_timecourse_N, 'k--', lw = 0.5, alpha = 0.4)


plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

plt.xlabel('time [s]')
plt.legend(['Visual Voxel', 'Auditory Voxel', 'Multimodal Voxel'])

plt.tight_layout()
```

## Analyzing our simulation's data

Now, in our simulated experiment we knew exactly when there were auditory and visual events, and we know the BOLD HRF. This means we can create expectations of what visual and auditory responses should look like in voxels we 'record' from (this is what the second of the four plots above shows).

We now use these expectations in a linear regression, using them as **explanatory variables**, or **regressors** in a **design matrix**. So, just like in the 3D scatterplot example above), the bold response in a single voxel (just like the `y_points` above, can depend on an `x1` and an `x2`. So, the analysis of fMRI data is the same as the analysis you've already seen. We know that for our present voxel simulation, `x1` and `x2` should look like the blue and green lines in the second figure above.

Let's see what this looks like in an analysis, starting with the multimodal voxel!

```python
# we first have to create a design matrix, by putting the different expectations together, and
# again adding the ones for fitting the intercept.
design_matrix = np.array([BOLD_timecourse_1, BOLD_timecourse_2, np.ones(BOLD_timecourse_2.shape[0])])

# first, we perform our analysis on the multimodal voxel.
# we again use the linear regression function
betas, _sse, _r, _svs = np.linalg.lstsq(design_matrix.T, multimodal_voxel_timecourse_N )

print('------------------------------------')
print('-----------Beta values:-------------')
print('------ Slope for Visual: %2.3f------'%betas[0])
print('----- Slope for Audition: %2.3f-----'%betas[1])
print('------- Intercept: %2.3f--------'%betas[2])
print('------------------------------------')


# now we also want to plot these
fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(411)

plt.title('Multimodal BOLD Signal With Neural and Scanner Noise')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         multimodal_voxel_timecourse_N, 'k--', lw = 0.5, alpha = 0.7)
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

ax = fig.add_subplot(412)
plt.title('With Fitted Intercept')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         multimodal_voxel_timecourse_N, 'k--', lw = 0.5, alpha = 0.7)
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
        np.ones(neural_signal_2.shape[0])*betas[2], lw = 1.5, color = 'r')
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

ax = fig.add_subplot(413)
plt.title('With Fitted Intercept and Visual Slope')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         multimodal_voxel_timecourse_N, 'k--', lw = 0.5, alpha = 0.7)
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
        np.ones(neural_signal_2.shape[0])*betas[2], lw = 1.5, color = 'r')
plt.plot(np.arange(neural_signal_1.shape[0])/20.0, 
         BOLD_timecourse_1*betas[0] + betas[2], lw = 2.5, color = 'b')
plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

ax = fig.add_subplot(414)
plt.title('With Fitted Intercept, Visual and Auditory Slope')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         multimodal_voxel_timecourse_N, 'k--', lw = 0.5, alpha = 0.3)
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
        np.ones(neural_signal_2.shape[0])*betas[2], lw = 1.5, color = 'r')
plt.plot(np.arange(neural_signal_1.shape[0])/20.0, 
         BOLD_timecourse_1*betas[0] + betas[2], lw = 2.5, color = 'b')
plt.plot(np.arange(neural_signal_2.shape[0])/20.0, 
         BOLD_timecourse_2*betas[1] + betas[2], lw = 2.5, color = 'g')

plt.axhline(0, c = 'k', linewidth = 0.75, alpha = 0.5)

plt.xlabel('time [s]')
plt.tight_layout()
```

```python
# we put the voxels' data together in one variable
# because GLM analysis can be done on many voxels at once
BOLD_data = np.array([visual_voxel_timecourse_N, auditory_voxel_timecourse_N, multimodal_voxel_timecourse_N])

# lstsq wants the data to be delivered in a specific direction, hence the .T there.
betas, _sse, _r, _svs = np.linalg.lstsq(design_matrix.T, BOLD_data.T )

fitted_visual_betas, fitted_auditory_betas, fitted_intercepts = betas

print 'RESULTS'
print 'look above to find the values used in the simulation'
print ''
print 'the different voxels: '
print '              [Visual\tAuditory\tMultimodal]'
print 'Beta Visual   ' + str(fitted_visual_betas)
print 'Beta Auditory ' + str(fitted_auditory_betas)
print ''
print 'intercepts, fitted and set:'
print '                       [Visual\tAuditory\tMultimodal]'
print 'Fitted intercepts      ' + str(fitted_intercepts)
print 'Intercepts, as defined ' + str(random_intercepts)
```
